```python
import pandas as pd
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

# __OCEANIAÂ´S COVID DATA__

## Data Table

```python
df = pd.read_excel("datasets/oceania_covid.xlsx",
                   sheet_name = "oceania_covid",
                   engine = "openpyxl",
                   #skiprows = range(20),
                   #skipfooter = 2
                  )
df.head()
```

```python
df.rename (columns = {"Country/Other" : "Country", "AreaName":"Continent", "RegName":"Region"},
           inplace = True
          )
```

```python
df['Total Recovered'] = df['Total Recovered'].replace(np.nan, 0)
df['Active Cases'] = df['Active Cases'].replace(np.nan, 0)
df['Total Tests'] = df['Total Tests'].replace(np.nan, 0)
df['Tests/ 1M pop'] = df['Tests/ 1M pop'].replace(np.nan, 0)
```

## Data Table with Renamed Columns and Remplaced NaNs (for 0s)

```python
df.head()
```

## Description of the variables

### *Categorical Variables*

__Country__  
*   This variable sets the country were the data was collected.



### *Numerical Variables*


__Total Cases__ 
 *  This variable sets the number of the total cases in the country.

__Total Recovered__  
*   This variable sets the total of the people that got well and didnt died.

__Active Cases__
 *  This variable sets how many cases of COVID are active.

__Tot Cases/1M pop__  
  * This variable sets how many cases of COVID has the country every million people.

__Deaths/1M pop__ 
 *  This variable sets how many deaths due to COVID has the country every million people.

__Total Tests__  
  * This variable sets how many COVID tests the country did.

__Tests/1M pop__  
 *  This variable sets how many COVID tests the country did every million people.

__Population__ 
  * This variable sets how many people has the country.

```python
x = df["Total Cases"]
y = df["Country"]

```

```python
mp.style.use("dark_background")
```

## Observations

 * We can conclude that Australia is the country with the most cases of covid, deaths from covid, and tests carried out.
    
 * Another interesting fact is that in small countries they have almost no tests, and they do not have many positive results either.
    
 * The country "Kiribati" only has 11 deaths.

## Graphics

```python
fig, ax = plt.subplots(figsize=(14, 5), layout='constrained')

plt.scatter(y, x)
plt.plot(y, x, )
plt.xlabel("Countries")
plt.ylabel("Total Cases (Millions)")
plt.title("Total Cases (Millions)", 
          fontdict={'family': 'serif', 
                    'color' : 'white',
                    'weight': 'bold',
                    'size': 18})
plt.show()
```

> This graphic shows the Total Cases of COVID in the previous countries, Australia has the highest number.

```python
x = df["Total Deaths"]
y = df["Country"]
```

```python
fig, ax = plt.subplots(figsize=(14, 5), layout='constrained')
plt.scatter(y, x)
plt.plot(y, x, )
plt.xlabel("Countries")
plt.ylabel("Total Deaths")
plt.title("Total Deaths", 
          fontdict={'family': 'serif', 
                    'color' : 'white',
                    'weight': 'bold',
                    'size': 18})
plt.show()
```

> This graphic shows the Total Deaths from COVID in the previous countries, Australia has the highest number.

```python
x = df["Total Tests"]
y = df["Country"]
```

```python
fig, ax = plt.subplots(figsize=(14, 5), layout='constrained')
plt.scatter(y, x)
plt.plot(y, x, )
plt.xlabel("Countries")
plt.ylabel("Total Tests")
plt.title("Total Tests", 
          fontdict={'family': 'serif', 
                    'color' : 'white',
                    'weight': 'bold',
                    'size': 18})
plt.show()
```

> This graphic shows the Total Tests from COVID in the previous countries Australia has the highest number.

```python
x = df["Total Cases"]
y = df["Country"]
x1 = df["Total Deaths"]
y1 = df["Country"]
x2 = df["Total Tests"]
y2 = df["Country"]
```

```python
fig, ax = plt.subplots(figsize=(14, 5), layout='constrained')


plt.scatter(y, x, color='white')
plt.scatter(y1, x1, color='blue')
plt.scatter(y2, x2, color='cyan')
plt.legend(["Total Cases", "Total Deaths", "Total Tests"])

plt.plot(y, x, color='white')
plt.plot(y1, x1,color='blue' )
plt.plot(y2, x2, color='cyan')
plt.xlabel("Countries")

plt.title("Total Tests vs Total Deaths vs Total Cases", 
          fontdict={'family': 'serif', 
                    'color' : 'white',
                    'weight': 'bold',
                    'size': 18})
plt.show()
```

> This graphic shows the Total Deaths vs Total Deaths vs Total Cases of COVID in the previous countries, Australia has the highest numbers of all the variables.
